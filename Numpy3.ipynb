```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n retsro DATA \n", sales_data[:,1:])
```

## Total sales per year

```python
x=np.sum(sales_data , axis=0)
x[1:]
```

```python
y=np.sum(sales_data[:,1:] , axis=0)
y
```

### Minimum sales per restuant

```python
min_sales=np.min(sales_data[:,1:], axis=1)
min_sales
```

## Maximum Sale per Year

```python
max_sales=np.max(sales_data[:,1:], axis=0)
max_sales
```

## Average Sales per Restruant

```python
avg_sales=np.mean(sales_data[:,1:], axis=1)
avg_sales
```

## Cumulative Sales

```python
cum_sales=np.cumsum(sales_data[:,1:], axis=1)
print(cum_sales)
print("#############")
print(np.mean(cum_sales, axis=0))
print("#############")
plt.figure(figsize=(10,6))
plt.plot(np.mean(cum_sales, axis=0))
plt.title("Average cumulative sales across across all restaurant" )
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
